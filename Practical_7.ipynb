### Exercise 1:
Create a simple line plot for `y = x` for `x` values ranging from 0 to 10.

### Exercise 14:
Plot a filled area under the curve of `y = x^2` for `x` ranging from 0 to 10.


```python
x = np.linspace(0, 10, 500)  
y = x ** 2  # y = x

# Create plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r-', linewidth=2, label='y = x')  
plt.fill_between(x, y, color='skyblue', alpha=0.4)  

plt.xlabel('x')
plt.ylabel('y')
plt.title('Filled Area Under y = x (0  x  10)')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercise 16:
Create a stacked bar chart for the following data: `Group A = [3, 5, 7]`, `Group B = [2, 6, 4]`, and `categories = ['X', 'Y', 'Z']`.


```python
categories = ['X', 'Y', 'Z']
group_a = [3, 5, 7]  
group_b = [2, 6, 4]  

plt.figure(figsize=(8, 6))
plt.bar(categories, group_a, label='Group A')
plt.bar(categories, group_b, bottom=group_a, label='Group B')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart (Group A + Group B)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.


```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
functions = [x, x**2, x**3, x**4]
titles = ['Linear', 'Quadratic', 'Cubic', 'Quartic']
styles = ['r-', 'g--', 'b-.', 'm:']

for ax, y, title, style in zip(axs.flat, functions, titles, styles):
    ax.plot(x, y, style)
    ax.set_title(title)
    ax.grid(True)
```

### Exercise 19:
Add error bars to a bar plot with `categories = ['A', 'B', 'C', 'D']` and `values = [3, 7, 5, 9]`, assuming the error is 1 for each value.

### Exercise 20:
Plot a sine wave using a dashed line and a cosine wave using a solid line on the same plot.


```python
x = np.linspace(0, 2 * np.pi, 100)
sine_y = np.sin(x)
cosine_y = np.cos(x)

plt.plot(x, sine_y, '--', label='Sine', color='blue')
plt.plot(x, cosine_y, '-', label='Cosine', color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Waves')
plt.legend()

plt.show()
```

### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.


```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute z values for the function z = sin(x) + cos(y)
Z = np.sin(X) + np.cos(Y)

# Create contour plot
plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(label='z = sin(x) + cos(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot of z = sin(x) + cos(y)')

# Show plot
plt.show()
```
