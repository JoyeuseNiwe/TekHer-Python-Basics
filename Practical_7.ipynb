### Exercise 21:

### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
values = [100, -20, 15, -10, 25, -5]
labels = ['Start', 'Loss', 'Gain', 'Loss', 'Gain', 'End']
cum_values = np.cumsum([0] + values[:-1])
colors = ['green' if v >= 0 else 'red' for v in values]
fig, ax = plt.subplots()
for i, (val, cum) in enumerate(zip(values, cum_values)):
    ax.bar(labels[i], val, bottom=cum, color=colors[i])

plt.title('Waterfall Chart')
plt.show()
```

### Exercise 49:

```python
labels = ['Speed', 'Strength', 'Agility', 'Endurance', 'Flexibility']
values = np.random.rand(5)
values = np.append(values, values[0])  # Close the loop

angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(subplot_kw={'polar': True})
ax.plot(angles, values, 'o-', linewidth=2)
ax.fill(angles, values, alpha=0.25)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
plt.title('Radar Chart of Attributes')
plt.show()
```
